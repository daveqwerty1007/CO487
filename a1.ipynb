```python
## a
from collections import Counter
import matplotlib.pyplot as plt
import os

def letter_frequency_from_file(file_path):
    with open(file_path, 'r') as file:
        text = file.read()

    cleaned_text = ''.join(filter(str.isalpha, text)).upper()

    frequency = Counter(cleaned_text)

    letters = sorted(frequency.keys())
    frequencies = [frequency[letter] for letter in letters]

    print(f'Frequency for file: {file_path}')
    for letter in range(ord('A'), ord('Z') + 1):
        print(f'{chr(letter)}: {frequency.get(chr(letter), 0)}')

    plt.bar(letters, frequencies)
    plt.xlabel('Letters')
    plt.ylabel('Frequency')
    plt.title(f'Letter Frequency in Ciphertext ({file_path})')
    plt.show()


cipher_dir = 'a1q1ciphertexts'  

for file_name in os.listdir(cipher_dir):
    if file_name.endswith(".txt"): 
        file_path = os.path.join(cipher_dir, file_name)
        letter_frequency_from_file(file_path)
```

```python
def generate_text_table(mapping):
    text_table = "Ciphertext Letter | Mapped Letter\n"
    text_table += "-" * 30 + "\n"
    for cipher_letter, mapped_letter in mapping.items():
        text_table += f"      {cipher_letter}         |      {mapped_letter}\n"
    return text_table


def substitute_letters(text, mapping):
    substituted_text = []
    for char in text:
        if char.upper() in mapping:
            if char.isupper():
                substituted_text.append(mapping[char.upper()])
            else:
                substituted_text.append(mapping[char.upper()].lower())
        else:
            substituted_text.append(char)
    return ''.join(substituted_text)

# Preliminary mapping for substitution cipher
preliminary_mapping = {
    'E': 'E', 'T': 'T', 'A': 'A', 'H': 'O', 'O': 'I', 'N': 'N', 'D': 'S', 'I': 'H', 
    'R': 'R', 'S': 'D', 'L': 'L', 'W': 'C', 'U': 'U', 'G': 'M', 'B': 'W', 
    'F': 'Y', 'P': 'G', 'Y': 'B', 'C': 'V', 'V': 'K', 'M': 'F', 'K': 'P', 
    'J': 'X', 'Q': 'J'
}

generate_text_table(preliminary_mapping)

substitute_letters(ciphertext_1, preliminary_mapping)
```

```python
from collections import defaultdict
import re

def kasiski_examination(ciphertext, min_length=3):
    sequences = defaultdict(list)
    for i in range(len(ciphertext) - min_length + 1):
        seq = ciphertext[i:i + min_length]
        if seq.isalpha():
            sequences[seq].append(i)
    gaps = []
    for seq, positions in sequences.items():
        if len(positions) > 1:
            for i in range(len(positions) - 1):
                gaps.append(positions[i + 1] - positions[i])
    
    return gaps
gaps = kasiski_examination(ciphertext_3)
print("Gaps between repeated sequences:", gaps)
```

```python
def factorize(n):
    """Returns the factors of a number n."""
    factors = []
    for i in range(2, n + 1):
        if n % i == 0:
            factors.append(i)
    return factors

def common_factors(gaps):
    """Finds the most common factors of the gaps."""
    all_factors = []
    for gap in gaps:
        all_factors.extend(factorize(gap))
    
    factor_counts = Counter(all_factors)
    return factor_counts.most_common()

gaps = [119, 783, 119, 7, 112, 665, 503, 818, 749, 168, 168, 651, 651, 50, 792, 161, 28, 252, 280, 560, 644, 763, 532, 322, 393, 217, 567, 770, 63, 420, 343, 553, 140, 21, 154, 286, 491, 70, 525, 133, 147, 196, 84, 14, 294, 616, 329, 245, 420, 149, 271, 420, 420, 147, 273, 147, 273, 147, 273, 301, 147, 273, 49, 49, 49, 49, 84, 343, 248, 312, 161, 420, 408, 602, 280, 140, 140, 280, 719, 686, 686, 124, 546, 231, 203, 28, 581, 441, 140, 441, 140, 441, 140, 441, 140, 441, 140, 441, 140, 441, 140, 161, 280, 140, 441, 140, 441, 140, 581, 595, 126, 294, 343, 448, 14, 462, 63, 231, 14, 189, 231, 203, 147, 595, 420, 112, 308, 420, 420, 420, 491, 294, 21, 273, 201, 413, 413, 413, 196, 462, 462, 462, 462, 462, 462, 224, 345, 154, 432, 343, 343, 15, 161, 196, 126, 328, 42, 35]
common_factors_of_gaps = common_factors(gaps)
print("Most common factors:", common_factors_of_gaps) 
```

```python
from collections import Counter

def frequency_analysis(text):
    frequencies = Counter(text)
    most_common = frequencies.most_common(1)[0][0] 
    shift = (ord(most_common) - ord('E')) % 26
    return shift

def decrypt_caesar(ciphertext, shift):
    decrypted = []
    for c in ciphertext:
        if c.isalpha():
            decrypted_char = chr((ord(c.upper()) - ord('A') - shift) % 26 + ord('A'))
            decrypted.append(decrypted_char)
        else:
            decrypted.append(c)
    return ''.join(decrypted)

def decrypt_vigenere_with_key_length(ciphertext, key_length):
    # Split the ciphertext into groups 
    groups = ['' for _ in range(key_length)]
    for i, char in enumerate(ciphertext):
        if char.isalpha():
            groups[i % key_length] += char
    key = []
    for group in groups:
        shift = frequency_analysis(group)
        key.append(chr(shift + ord('A')))
    
    print("Estimated key:", ''.join(key))
    return key

estimated_key = decrypt_vigenere_with_key_length(ciphertext_3, 7)
```

```python
def decrypt_vigenere(ciphertext, key):
    key_repeated = (key * (len(ciphertext) // len(key) + 1))[:len(ciphertext)]
    decrypted_text = []
    
    for c, k in zip(ciphertext, key_repeated):
        if c.isalpha():
            shift = ord(k.upper()) - ord('A')
            decrypted_char = chr((ord(c.upper()) - ord('A') - shift) % 26 + ord('A'))
            decrypted_text.append(decrypted_char)
        else:
            decrypted_text.append(c)

    return ''.join(decrypted_text)

key = "QCBZZST"
decrypted_message = decrypt_vigenere(ciphertext_3, key)
print("Decrypted message:", decrypted_message)
```
